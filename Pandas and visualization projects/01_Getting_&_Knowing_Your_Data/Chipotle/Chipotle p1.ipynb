### Step 9. Which was the most-ordered item?

```python
chipo['item_name'].value_counts().sort_values(ascending=False).head(1)
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
max_ordered_quantity = chipo.groupby('item_name')['quantity'].sum().max()
max_ordered_quantity
```

### Step 11. What was the most ordered item in the choice_description column?

```python
most_ordered_item = chipo.groupby('choice_description')['quantity'].sum().idxmax()
most_ordered_item
```

### Step 12. How many items were orderd in total?

```python
items_were_orderd = chipo['quantity'].sum()
items_were_orderd
```
