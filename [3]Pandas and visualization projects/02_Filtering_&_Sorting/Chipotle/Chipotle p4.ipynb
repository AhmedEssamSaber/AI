### Step 1. Import the necessary libraries

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv",sep = '\t')
df
```

### Step 4. How many products cost more than $10.00?

```python
convert_to_float = lambda x: float(x.replace('$', ''))
```

```python
count = len(df[df["item_price"] > 10.00])

print("Number of products costing more than $10.00:", count)
```

### Step 6. Sort by the name of the item

```python
sort = df.sort_values(by="item_name")
sort
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
most_expensive_item = df[df["item_price"] == df["item_price"].max()]

quantity_of_it = most_expensive_item["quantity"].values[0]

print("the quantity of the most expensive item ordered : ",quantity_of_it)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
How_many_times_was_a_Veggie_Salad_Bowl_ordered = len(df[df["item_name"] == "Veggie Salad Bowl"])
print("How many times was a Veggie Salad Bowl ordered : ",How_many_times_was_a_Veggie_Salad_Bowl_ordered)
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
How_many_time_did_someone_order_more_than_one_canned_soda = len(df[(df["item_name"] == "Canned Soda") & (df["quantity"] > 1)])
print("How many times did someone order more than one Canned Soda : ",How_many_time_did_someone_order_more_than_one_canned_soda)
```
