#### 41. How to sum a small array faster than np.sum? (★★☆)

````python
from numba import jit
import numpy as np

@jit(nopython=True)
def custom_sum(arr):
    result = 0.0
    for val in arr:
        result += val
    return result

arr = np.array([1.0, 2.0, 3.0])
result = custom_sum(arr)
print(result)
````

#### 47. Given two arrays, X and Y, construct the Cauchy matrix C (Cij =1/(xi - yj)) (★★☆)

````python
X = np.array([1.0, 2.0, 3.0])
Y = np.array([0.0, 2.5, 4.0])

num_x = len(X)
num_y = len(Y)
C = np.empty((num_x, num_y))

for i in range(num_x):
    for j in range(num_y):
        C[i, j] = 1.0 / (X[i] - Y[j])

print(C)
````

#### 50. How to find the closest value (to a given scalar) in a vector? (★★☆)

````python
x = np.array([1.6,2.8,3.5,4.3,5.9,6.6,7.5,10.1])

scalar = 6.0

closest_greater = float('inf')
closest_less = float('-inf')

for i in range(len(x)):

  if x[i] == scalar:
    print(f"the closest value is {x[i]}")
    break

  elif x[i] > scalar and x[i] < closest_greater:
    closest_greater = x[i]

  elif x[i] < scalar and x[i] > closest_less:
        closest_less = x[i]

if ((closest_greater < float('inf')) - scalar ) < ((closest_less > float('-inf')) - scalar ) :
    print(f"The closest value greater than scalar is {closest_greater}")

else:
    print(f"The closest value less than scalar is {closest_less}")
````

#### 63. Create an array class that has a name attribute (★★☆)

````python
class NamedArray(np.ndarray):
    def __new__(cls, input_array, name="Unnamed Array"):
        obj = np.asarray(input_array).view(cls)
        obj.name = name
        return obj

    def __array_finalize__(self, obj):
        if obj is None:
            return
        self.name = getattr(obj, 'name', 'Unnamed Array')

data = [1, 2, 3, 4, 5]
named_array = NamedArray(data, name="My Named Array")

print("Array Name:", named_array.name)

print("Sum:", np.sum(named_array))
````

#### 64. Consider a given vector, how to add 1 to each element indexed by a second vector (be careful with repeated indices)? (★★★)
